```python
e_scooter_daily_rides["Date"] = pd.to_datetime(e_scooter_daily_rides["Date"])
```

```python
cykelpassager_citysnittet["Date"] = pd.to_datetime(cykelpassager_citysnittet["Date"])
```

```python
e_scooters_in_traffic["Date"] = pd.to_datetime(e_scooters_in_traffic["Date"])
```

```python
e_scooters_in_traffic["Value"] = e_scooters_in_traffic["Value"].astype("int64")
```

```python
e_scooters_in_traffic["Measure area"] =  e_scooters_in_traffic["Measure area"].replace({"Antal fordon manuellt":"Antal fordon"})
```

```python
e_scooters_in_daily_traffic = e_scooters_in_traffic.loc[e_scooters_in_traffic["Measure area"] == "Resor per dag"]
```

```python
e_scooters_in_traffic = e_scooters_in_traffic.drop(e_scooters_in_traffic.loc[e_scooters_in_traffic["Measure area"] == "Resor per dag"].index)
```

```python
e_scooters_in_daily_traffic["Measure area"].unique()
```

```python
e_scooters_in_daily_traffic.info()
```

```python
e_scooters_in_traffic
```

```python
monthly_sums = (
    e_scooters_in_traffic
    .groupby([e_scooters_in_traffic["Date"].dt.to_period("M"),
              e_scooters_in_traffic["Measure area"]])["Value"]
    .median()
    .reset_index()
)

monthly_sums
```

```python
monthly_sums.to_excel(
    r"C:\Users\sugan\OneDrive\Desktop\Hyper_Island\e_scooters\output\Antal_fordon.xlsx",
    index=False
)
```

```python
monthly_sums_Dec = monthly_sums[monthly_sums["Date"].dt.month == 12]
monthly_sums_Dec
```
